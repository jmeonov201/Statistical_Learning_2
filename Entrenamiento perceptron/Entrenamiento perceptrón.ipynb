```python
def entrenamiento(X,etiqueta,pesos,LR):
    
    accuraci = 0
    
    while accuraci != 1:
        for i in range(200):
            
            #y_hat = np.matmul(X[i].reshape(1,3),pesos)
            
            if etiqueta[i] == 1 and np.matmul(X[i].reshape(1,3),pesos) < 0:
                
                pesos = pesos + (LR * X[i]).reshape(3,1) 
            
            elif etiqueta[i] == 0 and np.matmul(X[i].reshape(1,3),pesos) >= 0:
                
                pesos = pesos - (LR * X[i]).reshape(3,1)
    
            #print("Celda: ", i)
            #print("Equis: ", X[i])
            #print("W: ", pesos)
        
        pronostico = np.matmul(X,pesos) >= 0
        accuraci = accuracy_score(etiqueta,pronostico, normalize = True)
        
        #print("***W final***: ", pesos)
        #print("***Pronostico***: ", pronostico)
        print("***Accuracy***: ", accuraci)
    
    print(" ")
    print("***W final****: ", pesos)
    

    xx = np.linspace(2.5,16,100)
    yy = -(pesos[0]/pesos[1])*xx - (pesos[2]/pesos[1])  

    fig  = plt.figure(figsize=(20,10))

    ax1 = fig.add_subplot(1,1,1)

    ax1.scatter(X[:,0],X[:,1])
    ax1.plot(xx,yy)

    plt.show()
    
    #return accuraci
```
